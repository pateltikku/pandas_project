# Copying filtered data to a new dataframe.
to ensure original dataframe remains raw for future use.

```python
lang_python = 'Python'
filterlang_df = df[
    (df['Country'] == 'India') &
    (df['MainBranch'] == 'I am a developer by profession') &
    (df['LanguageWorkedWith'].str.contains(lang_python, na=False)) &
    (df['CurrencyDesc'] == 'Indian rupee') &
    (df['CompFreq'].notnull())  
].copy()
filterlang_df
```

```python
null_values = filterlang_df['CompFreq'].isnull().sum()
null_values
```

```python
null_values = filterlang_df['CompTotal'].isnull().sum()
null_values
```

```python
null_values = filterlang_df['ConvertedComp'].isnull().sum()
null_values
```

# Filling 'Nan' values in columns 'CompTotal' and 'ConvertedComp'
the mean of columns are used to replace 'NaN' values. 
*CompTotal- yearly salary before taxes.
*ConvertedComp- yearly salary in USD (as exchange rate on-19 feb 2020

```python
filterlang_df['CompTotal'].fillna(filterlang_df['CompTotal'].mean(), inplace=True)
filterlang_df['ConvertedComp'].fillna(filterlang_df['ConvertedComp'].mean(), inplace=True)
filterlang_df
```

# Formatting the numbers using apply and lambda.

```python
filterlang_df['CompTotal'] = filterlang_df['CompTotal'].apply(lambda x: "{:,.0f}".format(x))
filterlang_df
```

```python
filterlang_df['ConvertedComp'] = filterlang_df['ConvertedComp'].apply(lambda x: "{:,.0f}".format(x))
filterlang_df
```

# Extracting relevant information.

```python
columns_to_keep = ['Respondent', 'MainBranch', 'Age','Gender','Trans', 'CompTotal', 'ConvertedComp','Country','EdLevel', 'UndergradMajor', 'LanguageWorkedWith','YearsCodePro']
sliced_df = filterlang_df[columns_to_keep]
sliced_df
```
